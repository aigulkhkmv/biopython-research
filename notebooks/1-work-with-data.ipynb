### 2.1 Read FASTA

````python
from Bio import SeqIO
````

````python
for seq_record in SeqIO.parse("../data/orchid.fasta", "fasta"):
    print(seq_record.id)
    print(repr(seq_record.seq))
    print(len(seq_record))
````

### 2.2. Read GBK from GenBank

````python
from Bio import SeqIO
````

````python
for seq_record in SeqIO.parse("../data/orchid.gbk", "genbank"):
    print(seq_record.id)
    print(repr(seq_record.seq))
    print(len(seq_record))
````
